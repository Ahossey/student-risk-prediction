```python
# Standard libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
sns.set_theme(style='whitegrid', palette='muted')

print("Imports successful!")
```

```python
# Load the Data:
# Define paths
RAW_DATA_PATH = '../data/raw/'

df_math = pd.read_csv(RAW_DATA_PATH + 'student-mat.csv', sep=';')
df_port = pd.read_csv(RAW_DATA_PATH + 'student-por.csv', sep=';')

print(f"Math dataset shape: {df_math.shape}")
print(f"Portuguese dataset shape: {df_port.shape}")
```

```python
#First Inspection
print(f"Rows: {df_math.shape[0]} | Columns: {df_math.shape[1]}")
print(f"\nColumn names:\n{list(df_math.columns)}")
print(f"\nFirst 10 rows:")

df_math.head(10)
```

```python
# Data Quality Check
def data_quality_report(df, name):
    """
    Generate a professional data quality summary.
    Every real project starts with this.
    """
    
    print(f"\033[1mDATA QUALITY REPORT: {name}\033[0m")
    
    print(f"Shape: {df.shape[0]} rows × {df.shape[1]} columns")
    print(f"\nMissing Values: {df.isnull().sum().sum()} total")
    print(f"Duplicate Rows: {df.duplicated().sum()}")
    print(f"\nData Types:")
    print(df.dtypes.value_counts())
    print(f"\nNumeric Summary:")
    return df.describe()

data_quality_report(df_math, "Student Math Performance")
```

```python
print(df_math['G3'].describe())
print(f"\nUnique values in G3: {sorted(df_math['G3'].unique())}")
```

```python
# How many students are at risk?
threshold = 10

df_math['at_risk'] = (df_math['G3'] < threshold).astype(int)

total = len(df_math)
at_risk_count = df_math['at_risk'].sum()
passing_count = total - at_risk_count

print(f"Total students:     {total}")
print(f"At risk (fail):     {at_risk_count} ({at_risk_count/total*100:.1f}%)")
print(f"Passing:            {passing_count} ({passing_count/total*100:.1f}%)")
```

```python
#Combine Both Subjects

# Add subject label before merging
df_math['subject'] = 'math'
df_port['subject'] = 'portuguese'

# Combine into one master dataset
df = pd.concat([df_math, df_port], ignore_index=True)

print(f"Combined dataset shape: {df.shape}")
print(f"Subject breakdown:\n{df['subject'].value_counts()}")
```

```python
# Create Target Variable

# Binary target: 1 = at risk (failing), 0 = passing
df['at_risk'] = (df['G3'] < 10).astype(int)

print(f"Class distribution:\n{df['at_risk'].value_counts()}")
print(f"\nAt-risk rate: {df['at_risk'].mean()*100:.1f}%")
```

```python
#Save to Processed Folder
# Save combined dataset — this becomes our working file going forward
PROCESSED_PATH = '../data/processed/'

df.to_csv(PROCESSED_PATH + 'student_combined.csv', index=False)

print(f"Dataset saved successfully to {PROCESSED_PATH}student_combined.csv")
print(f"Final shape: {df.shape}")
```
